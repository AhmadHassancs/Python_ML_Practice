```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([0, 1, 2, 3, 4, 5, 7, 9])
ypoints = np.array([0, 3, 4, 5, 9, 7, 9 , 11])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 9, 4, 6, 1])
print(xpoints.shape)

plt.plot(xpoints)
plt.show()
```

```python
xpoints = np.array([1, 9, 4, 6, 1])
print(xpoints.shape)

plt.plot(xpoints, marker='p')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, '*-.g')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:k', ms = 5)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 40, mec = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
pi = '#f3889e'
plt.plot(ypoints, marker = 'o', ms = 20, mfc = pi , mec = 'r')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('Solarize_Light2')
   
data = {'unemployment_rate': [6.1, 5.8, 5.7, 5.7, 5.8, 5.6, 5.5, 5.3, 5.2, 5.2],
        'index_price': [1500, 1520, 1525, 1523, 1515, 1540, 1545, 1560, 1555, 1565]
       }
  
df = pd.DataFrame(data)
df.plot(x='unemployment_rate', y='index_price', kind='scatter')
plt.show()
```
